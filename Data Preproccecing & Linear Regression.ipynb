## Outliers

```python
# Function Outliers
def outliers(df):
    num_var = list(df._get_numeric_data().columns)
    for col_names in num_var:
        df[col_names] = df[col_names].apply(lambda y: df[col_names].mean()-3*df[col_names].std() 
                            if y < df[col_names].mean()-3*df[col_names].std() else y)
        df[col_names] = df[col_names].apply(lambda y: df[col_names].mean()+3*df[col_names].std() 
                            if y > df[col_names].mean()+3*df[col_names].std() else y)
    return(df)
```

```python
# Outliers
df = outliers(df)
df.describe()
```

## Encoding

```python
# Function Encoding
def encoding_char(x):
    char_var = list(set(x.columns) - set(x._get_numeric_data().columns))
    for col_names in char_var:
        f = pd.factorize(x[col_names])
        x[col_names] = pd.factorize(x[col_names])[0]
    return(x)
```

```python
# Encoding categorical data
df = encoding_char(df)
df
```

## Exporting

```python
# Exporting the dataset
df.to_csv('.csv', sep=';', index=False)
```

```python
# Splitting the dataset into the Training set and Test set

f = ['symboling','fueltype','aspiration','doornumber','carbody','drivewheel','enginelocation','wheelbase','carlength','carwidth','carheight','curbweight','enginetype','cylindernumber','enginesize','boreratio','stroke','compressionratio','horsepower','peakrpm','citympg','highwaympg']
X = df[f]
y = df['price']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
# В качестве у и х были выбраны переменные price и enginesize, а не compressionratio и fueltype, так как в них больше различных значений
```

## Simple Linear Regression

```python
# Fitting Simple Linear Regression to the Training set 
f1 = ['enginesize']
from sklearn.linear_model import LinearRegression
sr = LinearRegression().fit(X_train[f1], y_train)

# Getting parameters
sr.coef_, sr.intercept_
```

```python
# Predicting the Test set results
y_pred = sr.predict(X_test[f1])

# Coefficient of determination R^2
sr.score(X_train[f1], y_train).round(3)
```

```python
# Mean squared error
from sklearn.metrics import mean_squared_error
mean_squared_error(y_train, sr.predict(X_train[f1]), squared=False).round(2), mean_squared_error(y_test, y_pred, squared=False).round(2)
```

```python
# Visualising the Test set results
plt.scatter(X_test['enginesize'], y_test, color = 'red')
plt.plot(X_test['enginesize'], sr.predict(X_test[f1]), 'bo')
plt.title('price vs enginesize (Test set)')
plt.xlabel('enginesize')
plt.ylabel('price')
plt.show()
```

## Polynomial Regression

```python
# Fitting Polynomial Regression to the dataset
from sklearn.preprocessing import PolynomialFeatures
X_train_p = PolynomialFeatures().fit_transform(X_train)
X_test_p = PolynomialFeatures().fit_transform(X_test)

pr = LinearRegression().fit(X_train_p, y_train)

# Getting parameters
pr.coef_, pr.intercept_
```

```python
# Predicting the Test set results
y_pred = pr.predict(X_test_p)

# Coefficient of determination R^2
pr.score(X_train_p, y_train).round(3)
```

```python
# Исходя из значений коэффициента детерминации всех трех моделей, эта модель является наилучшей, так как ее R^2 больше чем у других моделец и равен 0,999, что практически равняется единице.
```

```python
# Visualising the Test set results
plt.scatter(X_test['enginesize'], y_test, color = 'red')
plt.plot(X_test['enginesize'], pr.predict(X_test_p), 'bo')
plt.title('price vs enginesize (Test set)')
plt.xlabel('enginesize')
plt.ylabel('price')
plt.show()
```
